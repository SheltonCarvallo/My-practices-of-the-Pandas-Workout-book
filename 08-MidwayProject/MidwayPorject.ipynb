### Create this pivot table again, only including countries in the OECD subset. In which of these countries does someone with an associate’s degree earn the most? In which of them does someone with a doctoral degree earn the most?

```python
path = '../pandas-workout-data/data/oecd_locations.csv'
oecd_countries = pd.read_csv(path, header=None, index_col=1, names=['abbrev', 'Country'])
oecd_countries
```

```python
# let's join these df
oecd_countries.join(
    pd.pivot_table(
    data=so_df,
    index='Country',
    columns='EdLevel',
    values='ConvertedCompYearly',
    aggfunc='mean'))
```

### In which of these countries does someone with an associate’s degree earn the most?

```python
(oecd_countries.join(
    pd.pivot_table(
    data=so_df,
    index='Country',
    columns='EdLevel',
    values='ConvertedCompYearly',
    aggfunc='mean'))['Associate degree (A.A., A.S., etc.)']
        .sort_values(ascending=False)
        .dropna()
        .apply(lambda x: f'{x:.2f}'))
```

### In which of them does someone with a doctoral degree earn the most?

```python
(
    oecd_countries.join(
        pd.pivot_table(
            data=so_df,
            index='Country',
            columns='EdLevel',
            values='ConvertedCompYearly',
            aggfunc='mean'
        )
    )['Other doctoral degree (Ph.D., Ed.D., etc.)']
    .dropna()
    .sort_values(ascending=False)
    .apply(lambda x: f'{x:.2f}')
)
```

### Replace the string value Less than 1 year in YearsCode with 0. Replace the string value More than 50 years with 51.

```python
so_df['YearsCode'].unique()
```

```python
# def is_less_or_more(value):
#     if value == 'More than 50 years':
#         return 51
#     elif value == 'Less than 1 year':
#         return 0
#     else:
#         return int(value)
    
```

```python
so_df['YearsCode'].apply(lambda x: 51 if x == 'More than 50 years' else(0 if x == 'Less than 1 year' else x))
```

```python
so_df['YearsCode'] = so_df['YearsCode'].apply(lambda x: 51 if x == 'More than 50 years' else(0 if x == 'Less than 1 year' else int(x)))
so_df['YearsCode'].unique()
```

### Create a new column in so_df, called experience, which will categorize the values in the YearsCode. Values can be
- Less than 1 year
- 1–2 years
- 3–5 years
- 6–10 years
- 11+ years

```python
so_df['YearsCode']
```

```python
bins = [0,1,3,6,11,100]
labels = ['Less than 1 year', '1-2 years', '3-5 years', '6-10 years', '11+ years']
pd.cut(x=so_df['YearsCode'], bins=bins, right=False, labels=labels)
```

```python
pd.cut(x=so_df['YearsCode'], bins=bins, right=False, labels=labels).value_counts()
```

```python
so_df['Experience'] = pd.cut(x=so_df['YearsCode'], bins=bins, right=False, labels=labels)
so_df['Experience'].value_counts()
```
