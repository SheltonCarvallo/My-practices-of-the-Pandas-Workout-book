```python
import pandas as pd
```

```python
path = '../../pandas-workout-data/data/oecd_locations.csv' 
df_oecd = pd.read_csv(filepath_or_buffer=path, header=None, names=['abbrev','country'], index_col='abbrev')
df_oecd
```

```python
path = '../../pandas-workout-data/data/oecd_tourism.csv'
columns_oecd_tourism = ['LOCATION', 'TIME', 'SUBJECT', 'Value']
oecd_tourism_df  = pd.read_csv(filepath_or_buffer=path, usecols=columns_oecd_tourism, index_col='LOCATION')
oecd_tourism_df 
```

```python
oecd_tourism_df  = oecd_tourism_df .loc[oecd_tourism_df ['SUBJECT'] == 'INT-EXP'].copy()
oecd_tourism_df  = oecd_tourism_df .drop('SUBJECT', axis=1)
```

```python
oecd_tourism_df 
```

Create a new series, ``tourism_spending``, in which the index reflects the country names (i.e., not abbreviations) and the value contains the average tourism spending for that country.

```python
series_mean_country = oecd_tourism_df.groupby('LOCATION')['Value'].mean()
series_mean_country.head(5)
```

```python
mean_tourism_spending = df_oecd.join(series_mean_country).reset_index().set_index('country')
del mean_tourism_spending['abbrev']
mean_tourism_spending
```

Create a third data frame, ``wine_df``, based on winemag-150k-reviews.csv. You only need two columns: ``country`` and ``points``.

```python
path = '../../pandas-workout-data/data/winemag-150k-reviews.csv'
columns = ['country', 'points']
wide_df = pd.read_csv(filepath_or_buffer=path, usecols=columns)
wide_df.head(5)
```

Get the mean wine score for each country, across all wine reviews, sorted in descending order.

```python
mean_wine_score_country = wide_df.groupby('country')['points'].mean().sort_values(ascending=False)
mean_wine_score_country.head(5)
# mean_wine_score_country.name = 'Value'
# mean_wine_score_country
```

Perform a standard join between the average wine scores per country and the average tourism spending per country. Where do you see NaN values? What do those NaN values mean?

Perform an outer join between the average wine scores per country and the average tourism spending per country. Where do you see NaN values? What do they mean now?

```python
df_3 = mean_tourism_spending.join(mean_wine_score_country, how='outer')
df_3
# The output frame’s index is the combination of the left’s index and the right’s index.
# We may thus end up with NaN values in columns from both the left and right, depending on which index value was missing
```

```python
df_3 = mean_wine_score_country.to_frame().join(mean_tourism_spending, how='outer')
df_3
```
