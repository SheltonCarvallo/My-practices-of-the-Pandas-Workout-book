```python
x = glob.glob('../../pandas-workout-data/data/*,*.csv')
x
# x[0].removeprefix('../pandas-workout-data/data\\').removesuffix('.csv').split(',')
```

```python
all_dfs = []
for path in glob.glob('../../pandas-workout-data/data/*,*.csv'):
    print(f'Reading csv from ==> {path}')
    city, state = path.removeprefix('../../pandas-workout-data/data\\').removesuffix('.csv').split(',')
    df = pd.read_csv(
        filepath_or_buffer=path,
        usecols=[0,1,2],
        header=0,
        names=['date_time', 'max_temp', 'min_temp']
    ).assign(
        city = city.replace('+',' ').title(),
        state = state.upper())
    all_dfs.append(df)
```

```python
df = pd.concat(all_dfs)
```

```python
df
```

Does the data for each city and state start and end at (roughly) the same time? How do you know?

```python
df.groupby(['state','city'])['date_time'].agg(["min","max"])
```

What is the lowest minimum temperature recorded for each city in the data set?

```python
df.groupby(['state', 'city'])['min_temp'].min().sort_values(ascending=True)
```

What is the highest maximum temperature recorded in each state in the data set?

```python
df.groupby('state')['max_temp'].max().sort_values(ascending=False)
```

What is the average difference in temperature (i.e., max – min) for each of the cities in our data set?

```python
df.groupby(['state', 'city'])[['min_temp', 'max_temp']].agg(['min','max'])
```

```python
df.groupby(['state', 'city'])[['min_temp', 'max_temp']].apply(lambda x: x.max() - x.min())
# for each series (column) takes the greater and the lowest and then perfomance the rest
```

```python
df.groupby(['state', 'city'])[['min_temp', 'max_temp']].apply(lambda x: np.mean(x.max() - x.min()))
# np.mean works with the values of a series an returns a scalar
# in this case  np.mean adds all the values in the state-cityy row an then devides for the amount of values
```
