### What team has won the most medals in all archery events?


- We’re interested in all years, so we specify slice(None) for the first index level.
- Archery is only a summer sport, so we can either indicate Summer for the second level or use slice(None).
- In the third level, we explicitly specify Archery so we only get rows for archery events.
- We ignore the fourth level, effectively making it a wildcard.

```python
df.loc[(slice(None), slice(None),'Archery')]
```

```python
df.loc[(slice(None), 'Summer', 'Archery')]
```

```python
df.dropna() # This remove every row whichs contains a NaN value
```

```python
df.dropna(subset='Medal') # This way we specifie a list of column names to consider when dropping rows
```

When subset is used, dropna() only considers missing values within these specified columns for dropping rows. Missing values in other columns are ignored.

```python
(
    df
    .dropna(subset=['Medal'])
    .loc[(slice(None), slice(None), 'Archery')])
```

```python
(
    df
    .dropna(subset=['Medal'])
    .loc[(slice(None), slice(None), 'Archery')]
    .value_counts()
)
```

```python
# Final answer
(
    df
    .dropna(subset=['Medal']) # get a new df
    .loc[(slice(None), slice(None), 'Archery'), 'Team']
    .value_counts()
)
```

### Going deep


As we have already seen, **loc** makes retrieving data from multi-indexed data frames pretty straightforward. However, sometimes we may want to use a multi-index differently. ``Pandas provides two other methods: xs and IndexSlice.``

Let’s start with **xs**, which lets us accomplish what we did in exercise 23: find matches for certain levels within a multi-index. For example, one question in the previous exercise asked you to find the mean height of participants in the “Table Tennis Women’s Team” event from all Olympics years. ``Using loc, we had to tell pandas to accept all values for Year, all values for Season, and all values for Sport—in other words, we only checked the fourth level of the multi-index``: the event. Our query looked like this:

```python
df.loc[(slice(None),
       'Summer',
       slice(None),
       "Table Tennis Women's Team"),
       'Height'
       ].mean()
```

Using **xs**, we can shorten that query to

```python
df.xs("Table Tennis Women's Team",
      level='Event'
      ).loc[:, 'Height']
```

```python
df.xs("Table Tennis Women's Team",
      level='Event'
      ).loc[:, 'Height'].mean() 
```

```python
df.xs(
    ('Summer', "Table Tennis Women's Team"),
    level=['Season', 'Event']
)
```

```python
df.xs(('Summer', "Table Tennis Women's Team"),
    level=['Season', 'Event']
).loc[:, 'Height']
```

```python
df.xs(('Summer', "Table Tennis Women's Team"),
    level=['Season', 'Event']
).loc[:, 'Height'].mean()
```

Notice that ``xs is a method`` and is thus invoked with round parentheses. By contrast, ``loc is an accessor attribute`` and is invoked with square brackets

A more general way to retrieve from a multi-index is **IndexSlice**. Remember when I mentioned earlier that we cannot use : inside round parentheses and thus need to say slice(None)? Well, ``IndexSlice solves that problem: it uses square brackets and can use slice syntax for any set of values. For example, we can say``

```python
df.loc[idx[1980:2016, :, 'Swimming':'Table tennis'], :]
```
