### What are the 10 most common words containing five or more letters in the wine descriptions? Turn all words into lowercase and remove all punctuation and symbols at the start or end of each word for easier comparison. Also remove the words flavors, aromas, finish, palate, and drink.

```python
symbols = ',$.?!$%'
```

```python
df['description'].loc[0]
```

```python
def remove_punctuation(text):
    lower_case_arr = text.lower().split()  
    new_arr = []
    for word in lower_case_arr:
       new_arr.append(word.strip(symbols))
    return new_arr
```

```python
def get_only_words_more_than_five(words):
    more_five_letters = []
    for word in words:
        if len(word) >= 5:
            more_five_letters.append(word)
    return more_five_letters
```

```python
def remove_wine_related_words(words):
    wine_related_word = ['aromas', 'flavors', 'finish', 'palate', 'drink']
    no_wine_related_word = []
    for word in words:
        if word not in wine_related_word:
            no_wine_related_word.append(word)
    return no_wine_related_word
```

```python
df['description_transformed'] = df['description'].apply(remove_punctuation).apply(get_only_words_more_than_five).apply(remove_wine_related_words)
```

```python
df['description_transformed'].loc[0]
```

```python
df['description_transformed'].loc[1]
```

```python
def get_serie_all_words(words_series):
    all_words = []
    for words in words_series:
        all_words += words
    return pd.Series(all_words)
        
```

```python
all_words = get_serie_all_words(df['description_transformed'])
all_words
```

```python
df = pd.read_csv(filepath_or_buffer=path, usecols=columns)
df
```

```python
def top_10_words(s):
    wine_related_word = ['aromas', 'flavors', 'finish', 'palate', 'drink']
    words = (
        s
        .str.lower()
        .str.split()
        .explode()
        .str.strip(symbols))
    
    return words.loc[(words.str.len() >= 5) & (~words.isin(wine_related_word))].value_counts().head(10)
```

```python
s = top_10_words(df['description'])
s
```
