### Count the number of vowels (a, e, i, o, and u) in each word. What is the average number of vowels per word?

```python
def counter_vowels(word):
    vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
    counter = 0
    for x in word:
        if x in vowels:
            counter += 1
    return counter        
```

```python
s.apply(counter_vowels)
```

```python
print(f'Average of vowels per word: {s.apply(counter_vowels).mean():,.2f}')
```
