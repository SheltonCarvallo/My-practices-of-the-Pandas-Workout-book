```python
import pandas as pd
```

```python
#nyc_taxi_2019-07

df_2019 = pd.read_csv(
    filepath_or_buffer='../../pandas-workout-data/data/nyc_taxi_2019-07.csv',
    usecols=['passenger_count', 'total_amount', 'payment_type']
    )
df_2019.head(5)
```

```python
df_2019['year'] = 2019
df_2019.head(5)
```

```python
df_2020 = pd.read_csv(
    filepath_or_buffer='../../pandas-workout-data/data/nyc_taxi_2020-07.csv',
    usecols=['passenger_count', 'total_amount', 'payment_type']
    )
df_2020.head(5)
```

```python
df_2020['year'] = 2020
df_2020.head(5)
```

```python
len(df_2019) + len(df_2020)
```

```python
df = pd.concat([df_2019, df_2020], axis=0)
df
```

How many rides were taken in 2019 and 2020, and what is the difference between these two figures?

```python
rides_2019 = len(df_2019)
rides_2020 = len(df_2020)
difference = rides_2019 - rides_2020
difference
```

How much money (in total) was collected in 2019 and 2020, and what was the difference between these two figures?

```python
amount_2019 = df.loc[df['year'] == 2019, 'total_amount'].sum()
amount_2020 = df.loc[df['year'] == 2020, 'total_amount'].sum()
print(f"2019: ${amount_2019:.2f} vs 2020: ${amount_2020:.2f}")
```

```python
print(f"Difference: ${(amount_2019 - amount_2020):.2f}")
```

Did the proportion of trips with more than one passenger change dramatically?

```python
trips_more_than_one_passenger_2019 = df.loc[(df['passenger_count'] > 1) & (df['year'] == 2019), 'passenger_count'].count()
trips_more_than_one_passenger_2019
```

```python
trips_more_than_one_passenger_2020 = df.loc[(df['passenger_count'] > 1) & (df['year'] == 2020), 'passenger_count'].count()
trips_more_than_one_passenger_2020
```

```python
df.loc[10] # This was to verify the values in index 10
```

Yes it changed dramatically

```python
print(f"Percetage: {((trips_more_than_one_passenger_2020/trips_more_than_one_passenger_2019)*100):.2f}%")
```

Did people use cash (i.e., payment_type of 2) less in 2020 than in 2019?

```python
cash_2019 = df.loc[(df['year'] == 2019) & (df['payment_type'] == 2), 'payment_type'].count()
cash_2019
```

```python
cash_2020 = df.loc[(df['year'] == 2020) & (df['payment_type'] == 2), 'payment_type'].count()
cash_2020
```
