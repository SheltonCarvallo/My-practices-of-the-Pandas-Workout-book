### What are the mean and median populations for these 1,000 largest cities? What does that tell you?

```python
df['population'].mean()
```

```python
df['population'].median() # This means a few big values (populations) pull the mean higher than the median
```

### Along these lines, if you remove the 50 most populous cities, what happens to the mean population? What happens to the median?

```python
df['population'].sort_values(ascending=False).head(50)
```

```python
top50_population_cities = df['population'].sort_values(ascending=False).head(50).index
```

```python
df = df.drop(index=top50_population_cities)
df
```

```python
df['population'].mean()
```

```python
df['population'].median() # Deleting the 50 largest cities by populations brings the median nearer to the mean, meaning there are no extreme population peaks.
```

### What is the northernmost city, and where does it rank?
