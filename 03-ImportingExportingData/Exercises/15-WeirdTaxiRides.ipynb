```python
import pandas as pd
import numpy as np
```

Load the CSV file into a data frame using only the four columns mentioned earlier: passenger_count, trip_distance, payment_type, and total_amount.

```python
# Just load passenger_count, trip_distance, payment_type, and total_amount.
df = pd.read_csv(
    filepath_or_buffer='../../pandas-workout-data/data/nyc_taxi_2019-01.csv',
    usecols=['passenger_count', 'trip_distance', 'payment_type', 'total_amount']
    )
df
```

How many taxi rides had more than eight passengers?

```python
taxi_rides_more_than_eight = df.loc[df['passenger_count']  > 8]
taxi_rides_more_than_eight['passenger_count'].count()
```

How many taxi rides had zero passengers?

```python
taxi_rides_zero_passengers = df.loc[df['passenger_count'] == 0]
taxi_rides_zero_passengers['passenger_count'].count()
```

How many taxi rides were paid for in cash and cost over $1,000?


```python
df.head(5)
```

```python
df['passenger_count'][(df['payment_type']== 2) & (df['total_amount'] > 1000)].count()
```

How many rides cost less than $0?

```python
df['passenger_count'][df['total_amount'] < 0].count()
```

How many rides traveled a below-average distance but cost an above-average amount?

```python
average_distance = df['trip_distance'].mean()
average_distance
```

```python
average_cost = df['total_amount'].mean()
average_cost
```

```python
df.loc[(df['trip_distance'] < average_distance) & (df['total_amount'] > average_cost)]
```

```python
df['passenger_count'][(df['trip_distance'] < average_distance) & (df['total_amount'] > average_cost)].count()
```
