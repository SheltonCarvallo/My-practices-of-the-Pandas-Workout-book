### Create a new column, trip_time, containing the amount of time each taxi ride took as a timedelta.

```python
df['trip_time'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
df['trip_time']
```

### Determine the number and percentage of rides that took less than 1 minute.

```python
one_minute = pd.to_timedelta('0 days 00:01:00')
one_minute
```

```python
df_rides_less_one_minute = df.loc[df['trip_time'] < one_minute]
df_rides_less_one_minute.shape[0]
```

```python
(df_rides_less_one_minute['trip_distance'].shape[0] / df['trip_time'].shape[0]) * 100
```

### Create a new column, trip_time_group, in which the values are short (< 10 minutes), medium (≥ between 10 minutes and 1 hour), and long (> 1 hour).

```python
df['trip_time'].max()
```

```python
times = ['00:00:00', '00:10:00', '01:00:00.001', df['trip_time'].max() + pd.Timedelta('00:00:01')]
```

```python
bins = pd.to_timedelta(times)
bins
```

```python
labels = ['short', 'medium', 'long']
```

```python
df['trip_time_group'] = pd.cut(x=df['trip_time'], bins=bins, right=False, labels=labels)
df['trip_time_group']
```

```python
df.loc[df['trip_time_group'].isnull(), 'trip_time']
```

### Determine the proportion of rides in each group.

```python
df['trip_time_group'].value_counts(normalize=True)
```

### For each value in trip_time_group, determine the average number of passengers.

```python
# obseved parameter This only applies if any of the groupers are Categoricals. If True: only show observed values for categorical groupers. 
# If False: show all values for categorical groupers.
df.groupby('trip_time_group', observed=False)['passenger_count'].mean()
```
