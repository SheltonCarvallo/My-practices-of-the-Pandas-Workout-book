```python
import pandas as pd
import numpy as np
```

```python
s = pd.read_csv(filepath_or_buffer='../../pandas-workout-data/data/nyc-temps.txt').squeeze()
s
```

```python
df = pd.DataFrame(data={
    'temp': s,
    'hour': [0,3,6,9,12,15,18,21] * 91
})
df
```

```python
df.loc[df['temp'] < 0, 'temp'] = 0
df
```

```python
df.describe()
```

### Beyond the exercise

Set all the odd temperatures to the mean of all temperatures

```python
df['temp'] = df['temp'].astype(np.float64)
df['temp']
```

```python
mean_temp = df['temp'].mean()
mean_temp
```

```python
df.loc[df['temp'] % 2 != 0]
```

```python
df.loc[df['temp'] % 2 != 0, 'temp'] = mean_temp
```

```python
df.loc[(df['temp'] % 2 == 0) & (df['hour'].isin([9 ,18])), 'temp'] = 3
df
```
