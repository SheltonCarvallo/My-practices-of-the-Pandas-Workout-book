```python
import pandas as pd
import numpy as np
```

```python
passenger_count = pd.read_csv(filepath_or_buffer='C:/Users/Shelton/Desktop/Programming/Pandas/Pandas-Cookbook/pandas-workout-data/data/taxi-passenger-count.csv',
                         header=None ).squeeze()
passenger_count.head(5)
```

```python
trip_distances = pd.read_csv(filepath_or_buffer='C:/Users/Shelton/Desktop/Programming/Pandas/Pandas-Cookbook/pandas-workout-data/data/taxi-distance.csv',
                         header=None ).squeeze()
trip_distances.head(5)
```

```python
df = pd.DataFrame({
    'trip_distance': trip_distances,
    'passenger_count': passenger_count
})
df
```

```python
# iqr = df['trip_distance'].quantile(0.75) - df['trip_distance'].quantile(0.25)
iqr = (
    df['trip_distance'].quantile(0.75) -
    df['trip_distance'].quantile(0.25)
)
iqr
```

```python
# outlier distances -- below
df[
    df['trip_distance'] < df['trip_distance'].quantile(0.25) -
    1.5 * iqr
]
```

```python
# outlier distances -- above
df[
    df['trip_distance'] > df['trip_distance'].quantile(0.75) +
    1.5 * iqr
]
```

```python
# bc there's no outliers in the below
df['passenger_count'][
    df['trip_distance'] > df['trip_distance'].quantile(0.75) +
    1.5 * iqr
].mean()
```
