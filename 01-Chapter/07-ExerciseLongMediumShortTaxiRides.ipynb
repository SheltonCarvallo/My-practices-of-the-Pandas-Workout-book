```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(filepath_or_buffer='C:/Users/Shelton/Desktop/Programming/Pandas/Pandas-Cookbook/pandas-workout-data/data/taxi-distance.csv', header=None)
data # Remember the read_csv() method always returns a DataFrame
```

```python
serie = data.squeeze()
serie
```

```python
category_serie = serie.astype(str)
category_serie
```

```python
category_serie.loc[serie <= 2] = 'Short'
category_serie.loc[(serie > 2) & (serie <= 10)] = 'Medium'
category_serie.loc[serie > 10] = 'Long'
category_serie
```

```python
category_serie.value_counts()
```

This certainly works, but as you probably guessed, there is a more efficient approach. The `pd.cut()` method allows us to set numeric boundaries.
Morover, ir can assign labels to each of the bins

```python
pd.cut(serie, 
       bins=[0,2,10, serie.max()],
       include_lowest=True,
       labels=['Short', 'Medium', 'Long'])
```

```python
pd.cut(serie,
       bins=[0,2,10,serie.max()],
       include_lowest=True,
       labels=['Short', 'Medium', 'Lonq']).value_counts()
```
