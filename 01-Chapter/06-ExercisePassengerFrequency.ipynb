Your task in this exercise is to show what percentage of taxi rides had only one passenger versus the (theoretical) maximum of six passengers.

```python
serie =  data_frame.squeeze()
serie
```

```python
ride_one_person = serie[serie == 1]
ride_one_person
```

```python
frequency_1_ride = ride_one_person.count()
frequency_1_ride
```

```python
ride_six_people = serie[serie == 6]
ride_six_people
```

```python
frequency_6_people = ride_six_people.count()
frequency_6_people
```

```python
total_rides = serie.count()
total_rides
```

```python
percentage_1_person_ride = round((frequency_1_ride/total_rides) * 100, 2)
percentage_6_people_ride = round((frequency_6_people/total_rides) * 100, 2)
print(f'Percentages: \n1 person for ride {percentage_1_person_ride} %\n6 peopple for ride {percentage_6_people_ride} %')
```

There’s nothing inherently wrong with doing things this way, but there’s a far easier technique

```python
serie.value_counts()
```

Because we get back a series from value_counts, we can use all our series tricks on it.

```python
serie.value_counts().loc[[1,6]]
```

But we’re interested in the percentages, not the raw values. Fortunately, **value_counts** `has an optional normalize parameter that returns the fraction if set to True`

```python
serie.value_counts(normalize=True).loc[[1,6]]
```

```python
serie.value_counts(normalize=True).loc[[1,6]] * 100
```
